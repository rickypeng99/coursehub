```python
def insert_course(cursor, data_course):
    add_course = ("INSERT INTO CourseHub.Courses "
               "(CRN, dept, idx, title, special_title, credit, term, type, session) "
               "VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)")
    cursor.execute(add_course, data_course)
    
def insert_user(cursor, data_user):
    try:
        add_user = ("INSERT INTO CourseHub.Users "
                   "(net_id, internal_point, major, user_name, password, first_name, middle_name, last_name, description) "
                   "VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)")
        cursor.execute(add_user, data_user)
    except:
        print("insert user failed")
        
def insert_user_skill(cursor, data_user_skill):
    add_user_skill = ("INSERT INTO CourseHub.Users_Skills "
               "(net_id, skill) "
               "VALUES (%s,%s)")
    cursor.execute(add_user_skill, data_user_skill)
    
def delete(table):
    cnx = mysql.connector.connect(host="localhost",
      user="root",
      passwd="UIUCBBKinG19981234")
    cursor = cnx.cursor()
    cursor.execute("DELETE FROM CourseHub.{} ".format(table))
    cnx.commit()
    cursor.close()
    cnx.close()
    
def random_skills():
    language_list = ["CSS","Html","SQL","PHP","Haskell","C++","C","Python","Java","C#","R","Java Script","Go","Swift","Ruby"]
    topic_list = ["Mathematical foundations",
                 "Algorithms",
                 "Data Structures",
                 "Artificial intelligence",
                 "Automated reasoning",
                 "Computer vision",
                 "Machine learning",
                 "Evolutionary computing",
                 "Natural language processing",
                 "Robotics",
                 "Networking",
                 "Computer security",
                 "Cryptography",
                 "Computer architecture",
                 "Operating systems",
                 "Computer graphics",
                 "Image processsing",
                 "Information visualization",
                 "Parallel computing",
                 "Concurrency",
                 "Distributed computing",
                 "Relatinal databases",
                 "Structured Storage",
                 "Data mining",
                 "Programming languages",
                 "Compilers",
                 "Scientific computing",
                 "Formal methods",
                 "Software engineering",
                 "Algorithm design",
                 "Computer programming",
                 "Human-computer interaction",
                 "Reverse engineering"]
    language_total = random.randrange(2,15)
    topic_total = random.randrange(3,8)
    output = np.append(np.random.choice(language_list, language_total, replace=False), np.random.choice(topic_list, topic_total, replace=False)).tolist()
    return output
    
    
def random_user(total):
    # [FirstName,LastName,NetID,Username,password,points]
    output = []
    netIds = random_net_id(total)
    for i in range(total):
        person = [names.get_first_name(), names.get_last_name(), netIds[i], netIds[i], ''.join(random.choice(string.ascii_letters + string.digits) for x in range(8)),random.randrange(0,1000)]
        output.append(person)
    return output

def random_net_id(total):
    output = []
    for i in range(total):
        if (i % 1000 == 0):
            print(i)
        chars = string.ascii_lowercase
        size = random.randrange(2,8)
        cur = ''.join(random.choice(chars) for x in range(size)) + str(random.randrange(1,9))
        while (cur in output):
            size = random.randrange(2,8)
            cur = ''.join(random.choice(chars) for x in range(size)) + str(random.randrange(1,9))
        output.append(cur)
    return output
        
```
